```python
## Splitting data between fraudulent and legit transactions
legit_data = data.loc[data["Class"] == 0]
fraudulent_data = data.loc[data["Class"] != 0]
```

```python
## Understanding descriptons of the splitted data 
print(legit_data["Amount"].describe(),fraudulent_data["Amount"].describe())
```

```python
## Random sampling 492 legit transcations with that of fraudulent transactions 

legit_sample = legit_data.sample(492)

## Concatenating the fraudulent transcations with legit samples 

new_data = pd.concat([legit_sample, fraudulent_data], axis = 0)


print(new_data.head())

new_data
```
